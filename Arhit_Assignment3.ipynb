```python
# Impute missing values for all numeric variables
data['price'] = data['price'].fillna(data['price'].mean())
data['area'] = data['area'].fillna(data['area'].mean())
data['bedrooms'] = data['bedrooms'].fillna(data['bedrooms'].mean())
data['bathrooms'] = data['bathrooms'].fillna(data['bathrooms'].mean())
data['stories'] = data['stories'].fillna(data['stories'].mean())
data['parking'] = data['parking'].fillna(data['parking'].mean())

# Impute missing values for categorical variables
data['mainroad'] = data['mainroad'].fillna(data['mainroad'].mode()[0])
data['guestroom'] = data['guestroom'].fillna(data['guestroom'].mode()[0])
data['basement'] = data['basement'].fillna(data['basement'].mode()[0])
data['hotwaterheating'] = data['hotwaterheating'].fillna(data['hotwaterheating'].mode()[0])
data['airconditioning'] = data['airconditioning'].fillna(data['airconditioning'].mode()[0])
data['furnishingstatus'] = data['furnishingstatus'].fillna(data['furnishingstatus'].mode()[0])

# Check for missing values
data.isnull().sum()
```

```python
# Find Outliers and replace them 
# Boxplot of Price
sns.boxplot(data=data, x='price')
plt.xlabel('Price')
plt.title('Boxplot of Price')
plt.show()
```

```python
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
```

```python

# Calculate skewness and kurtosis
skewness = numeric_columns.skew()
kurtosis = numeric_columns.kurtosis()

# Define a function to replace outliers with the median value
def replace_outliers(column):
    median = column.median()
    std = column.std()
    outliers = (column - median).abs() > 3 * std
    column[outliers] = median
    return column
    
# Replace outliers in each numeric column with the median value
for column in numeric_columns:
    numeric_columns[column] = replace_outliers(numeric_columns[column])

# Update the original DataFrame with the replaced values
data.update(numeric_columns)

# Verify the replaced outliers
replaced_outliers = pd.DataFrame({'Skewness': numeric_columns.skew(), 'Kurtosis': numeric_columns.kurtosis()})
print("Replaced Outliers:")
print(replaced_outliers)
```

```python
# Encode categorical variables
categorical_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'furnishingstatus']
encoded_data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
X = encoded_data.drop('price', axis=1)  # Independent variables
y = encoded_data['price'] 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
linear_reg = LinearRegression()
decision_tree_reg = DecisionTreeRegressor()
random_forest_reg = RandomForestRegressor()
lasso_reg = Lasso()
ridge_reg = Ridge()

models = [linear_reg, decision_tree_reg, random_forest_reg, lasso_reg, ridge_reg]
```

```python
# Define a function to calculate the RMSE and R2 score
def calculate_metrics(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    return rmse, r2
```

```python
# Train and test each model
for model in models:
    rmse, r2 = calculate_metrics(model)
    print(type(model).__name__)
    print('RMSE:', rmse)
    print('R2 Score:', r2)
    print()
```

```python
# Test each model with cross-validation
from sklearn.model_selection import cross_val_score

for model in models:
    scores = cross_val_score(model, X_scaled, y, cv=5, scoring='r2')
    print(type(model).__name__)
    print('R2 Scores:', scores)
    print('Mean R2 Score:', np.mean(scores))
    print()
```

```python
# plot the RMSE and R2 score for each model
rmse_scores = []
r2_scores = []
for model in models:
    rmse, r2 = calculate_metrics(model)
    rmse_scores.append(rmse)
    r2_scores.append(r2)

plt.figure(figsize=(10, 5))
plt.title('RMSE Score Comparison')
plt.xlabel('Model')
plt.ylabel('RMSE Score')
plt.bar([type(model).__name__ for model in models], rmse_scores)
plt.show()

plt.figure(figsize=(10, 5))
plt.title('R2 Score Comparison')
plt.xlabel('Model')
plt.ylabel('R2 Score')
plt.bar([type(model).__name__ for model in models], r2_scores)
plt.show()

    
```
