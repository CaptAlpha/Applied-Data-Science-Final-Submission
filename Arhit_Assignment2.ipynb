6. Find the outliers and replace the outliers


```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
```

```python

# Create a box plot for each numeric column
plt.figure(figsize=(10, 6))
sns.boxplot(data=numeric_columns)
plt.title("Box Plot of Numeric Columns")
plt.xticks(rotation=45)
plt.show()
```

```python


# Calculate skewness and kurtosis
skewness = numeric_columns.skew()
kurtosis = numeric_columns.kurtosis()

# Define a function to replace outliers with the median value
def replace_outliers(column):
    median = column.median()
    std = column.std()
    outliers = (column - median).abs() > 3 * std
    column[outliers] = median
    return column
    
# Replace outliers in each numeric column with the median value
for column in numeric_columns:
    numeric_columns[column] = replace_outliers(numeric_columns[column])

# Update the original DataFrame with the replaced values
df.update(numeric_columns)

# Verify the replaced outliers
replaced_outliers = pd.DataFrame({'Skewness': numeric_columns.skew(), 'Kurtosis': numeric_columns.kurtosis()})
print("Replaced Outliers:")
print(replaced_outliers)
```

7. Check for Categorical columns and perform encoding.


```python
# Check categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
# use the value_counts() function to display the unique values for each categorical feature
for column in categorical_cols:
    print(df[column].value_counts())
```

```python
# Use dummy variables to convert categorical features to numeric features

for column in categorical_cols:
    df = pd.concat([df, pd.get_dummies(df[column], prefix=column)], axis=1)
    df.drop(column, axis=1, inplace=True)

df.head()

# Replace True and False with 1 and 0
df = df.replace(True, 1)
df = df.replace(False, 0)

df.head()
```
